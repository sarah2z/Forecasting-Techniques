```python
#importing the covid19 confirmed dataset

confirmed_cases = pd.read_csv("time_series_covid19_confirmed_global.csv")
confirmed_cases.head()
```

```python
#aggregating the rows by Country
confirmed_cases_aggregated = confirmed_cases.groupby("Country/Region").sum()
confirmed_cases_aggregated.head(5)
```

```python
confirmed_cases_aggregated.shape #195 countries, 630 days
```

```python
sns.set_palette(sns.color_palette("husl", 8))
fig = plt.figure(figsize = (16,9))
# iloc = index location. We are taking the top 5 countries. We specify the columns first and then the rows
plt.title("Top 5 countries with most confirmed cases", size = 25)
ax = sns.barplot(data = top_5_confirmed_cases.iloc[:5], y = "total_confirmedcases", x = "Country/Region", linewidth = 2, edgecolor = 'black',)
plt.xlabel("Country/Region")
plt.ylabel("Total Confirmed Cases")
plt.show()
```

```python
# Top 5 countries based on death cases
# reset_index adds indices to the data
top_5_death_cases = corona_data.groupby(by = 'Country/Region').max()[['total_deathcases']].sort_values(by = ['total_deathcases'], ascending = False).reset_index()
top_5_death_cases.head(5)
```

```python
sns.set_palette(sns.color_palette("Paired"))
fig = plt.figure(figsize = (16,9))
# iloc = index location. We are taking the top 5 countries. We specify the columns first and then the rows
plt.title("Top 5 countries with most death cases", size = 25)
ax = sns.barplot(data = top_5_death_cases.iloc[:5], y = "total_deathcases", x = "Country/Region", linewidth = 2, edgecolor = 'black',)
plt.xlabel("Country/Region")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
#To identify the number of active cases for each country given by the formula:
#No of active cases = confirmed – (deaths + recovered) for each country

# Create a new column named 'Active Cases'
corona_data['Active_Cases'] = corona_data['total_confirmedcases'] - (corona_data['total_recoveredcases'] + 
                                                                     corona_data['total_deathcases'])
corona_data.head()
```

```python
# Calculate the recovery rate = Total no of recovered cases/ Total no of confirmed cases *100
corona_data["Recovery_Rate"] = (corona_data["total_recoveredcases"]/corona_data["total_confirmedcases"])*100
```

```python
corona_data["Recovery_Rate"]=round(corona_data["Recovery_Rate"],2)
corona_data.head()
```

```python
# Mortality Rate = Total no of deaths/ total no of confirmed * 100
corona_data["Mortality_Rate"] = (corona_data["total_deathcases"]/corona_data["total_confirmedcases"])*100
corona_data.head()
```

```python
# Top 5 countries based on active cases
# reset_index adds indices to the data
top_5_active_cases = corona_data.groupby(by = 'Country/Region').max()[['Active_Cases']].sort_values(by = ['Active_Cases'], ascending = False).reset_index()
top_5_active_cases.head(6)
```

```python
sns.set_palette(sns.color_palette("husl", 20))
fig = plt.figure(figsize = (16,9))
# iloc = index location. We are taking the top 5 countries. We specify the columns first and then the rows
plt.title("Top 5 countries with most Active cases", size = 25)
ax = sns.barplot(data = top_5_active_cases.iloc[:5], y = "Active_Cases", x = "Country/Region", linewidth = 2, edgecolor = 'black',)
plt.xlabel("Country/Region")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#Finding the highest value for each of the columns
def give_largest(col,n):
    largest = col.nlargest(n).reset_index(drop = True)
    data = [x for x in largest]
    index = [f'{i}_largest' for i in range(1,len(largest)+1)]
    return pd.Series(data,index=index)


def n_largest(corona_data, axis, n):
    '''
    Function to return the n-largest value of each
    column/row of the input DataFrame.
    '''
    return corona_data.apply(give_largest, axis = axis, n = n)


n_largest(corona_data,axis = 0, n = 5)
```

```python
#importing the vaccinated per country dataset
vaccine_df = pd.read_csv("Vaccinated per country.csv")
vaccine_df.head(100)
```

```python
#dropping the useless columns
useless_cols = ["iso_code", "date", "total_vaccinations","people_fully_vaccinated",
                "total_boosters","daily_vaccinations_raw","total_vaccinations_per_hundred","people_vaccinated_per_hundred","people_fully_vaccinated_per_hundred",
                "total_boosters_per_hundred","daily_vaccinations_per_million"]
```

```python
vaccine_df.drop(useless_cols,axis=1,inplace=True)
```

```python
vaccine_df.head(100)
```

```python
# Finding the sum of missing values for each column
vaccine_df.isnull().sum()
```

```python
#Dropping the first row to remove values '##country+name' and '#total+vaccinations+daily'
vaccine_df1=vaccine_df.iloc[1:,:]
vaccine_df1.set_index('location')
```

```python
#Replacing the empty values for column 'people_vaccinated' with 0 and converting the column type from object to int
vaccine_df1['people_vaccinated'] = vaccine_df1['people_vaccinated'].fillna(0.0).astype(int)
vaccine_df1
```

```python
# Rename the 'location' column to 'Country/Region'
vaccine_df1.rename(columns = {"location":"Country/Region"}, inplace=True)
vaccine_df1
```

```python
#aggregating the column ‘People_Vaccinated’ by the max value grouped by each region 
vaccine_df1_aggregated=vaccine_df1.groupby('Country/Region')['people_vaccinated'].max()
vaccine_df1_aggregated=pd.DataFrame(vaccine_df1_aggregated)
```

```python
vaccine_df1_aggregated.head()
```

```python
# Top 10 countries/regions with the most vaccinated people
# reset_index adds indices to the data
top_10_vaccinated_regions = vaccine_df1_aggregated.groupby(by = 'Country/Region').max()[['people_vaccinated']].sort_values(by = ['people_vaccinated'], ascending = False).reset_index()
top_10_vaccinated_regions
```

```python
#Countries/regions with the most no of vaccinated individuals
sns.set_palette(sns.color_palette("husl", 10))
fig = plt.figure(figsize = (18,5))
plt.title("Top 10 countries/regions with most no of vaccinated individuals", size = 25)
ax = sns.barplot(data = top_10_vaccinated_regions[top_10_vaccinated_regions['Country/Region'].isin(['China', 'India', 'United States', 'Brazil', 'Indonesia','Japan','Mexico','Pakistan','Germany','Turkey'])], x = 'Country/Region', y ='people_vaccinated',linewidth = 2, edgecolor = 'black')
plt.xlabel("Regions")
plt.ylabel("Total no of Vaccinations")
plt.show()
```

```python
#countries/regions with the least vaccinated people

least_vaccinated_regions = vaccine_df1_aggregated.groupby(by = 'Country/Region').min()[['people_vaccinated']].sort_values(by = ['people_vaccinated'], ascending = True).reset_index()
least_vaccinated_regions.head(5)
```

```python
sns.set_palette(sns.color_palette("pastel", 75))
fig = plt.figure(figsize = (18,5))
plt.title("Countries/regions with least no of vaccinated individuals", size = 25)
ax = sns.barplot(data = least_vaccinated_regions.iloc[:5], y = "people_vaccinated", x = "Country/Region", linewidth = 2, edgecolor = 'black',)
plt.xlabel("Regions")
plt.ylabel("Total no of Vaccinations")
plt.show()
```
